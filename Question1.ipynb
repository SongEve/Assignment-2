```python
from numpy import arange
import numpy as np
from matplotlib import pyplot
```

```python
alpha = 0.001
beta1 = 0.9
beta2 = 0.999
learning_rate = 0.3
```

```python
#objective function
def grad(x,h):
    if x<1:
        return -1
    elif 1<x<(1+h):
        return 1
    elif (1+h)<x<(1+2*h):
        return -1
```

```python
def max_h(grad, n,learning_rate, beta1,beta2,epsilon=0):
    for h in np.arange(0.3,1,alpha):
        m = 0
        v = 0
        x = 0
        for t in range(1,n):
            g = grad(x,h)
            m = beta1 * m + (1-beta1)*g
            v = beta2 * v + (1-beta2)*g**2
            mhat = m/(1-beta1**t)
            vhat = v/(1-beta2**t)
            x = x-learning_rate*mhat / (vhat**0.5+epsilon)
            
            if x > (1+h):
                print(x,h)
                break
        if x < (1+h):
            print("Maximum hight=",h)
            break
max_h(grad,100,learning_rate,beta1,beta2)
```
